```python
#Preencher nomes das marcas
marcas = [
    "Samsung", "Apple", "Motorola", "Xiaomi", "Realme",
    "Asus", "Nokia", "Lenovo", "LG", "Sony", "Huawei", "Vivo", "Redmi", "Poco", "Multilaser", "Infinix"
    "Honor", "Zte", "CAT", "Blu", "Cubot", "Positivo", "Nubia", "Multi", "Vovofone", "Oppo", "Itel", "Redmagic", "Blackberry"
]

def extrair_marca(nome):
    if pd.isna(nome):
        return None
    
    nome_lower = nome.lower()

    if "iphone" in nome_lower:
        return "Apple"
    
    if "honor" in nome_lower:
        return "Honor"
    
    if "moto" in nome_lower:
        return "Motorola"

    for marca in marcas:
        if marca.lower() in nome_lower:
            return marca
    return None
```

```python
mask = df['brand'].isna()

df.loc[mask, 'brand'] = df.loc[mask, 'name'].apply(extrair_marca)
```

```python
df = df.dropna(subset=['brand']).reset_index(drop=True)
```
